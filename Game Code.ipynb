```python
import pygame, sys, random

pygame.mixer.pre_init(44100, -16, 2, 512)
pygame.init()
Block_height = 30
Block_width = 20
FRUIT = pygame.image.load("Fruit.png")
BG = pygame.mixer.Sound("Snake_BGM.mpeg")
FONT = pygame.font.Font("UrbanJungleDEMO.otf", 20)
BG.play()
class Snake:
    def __init__(self):
        self.body = [pygame.math.Vector2(5, 10), pygame.math.Vector2(4, 10), pygame.math.Vector2(3, 10)]
        self.direction = pygame.math.Vector2(0, 0)
        self.new_block = False
        
        self.snake_up = pygame.image.load('snake_up.png').convert_alpha()
        self.snake_down = pygame.image.load('snake_down.png').convert_alpha()
        self.snake_right = pygame.image.load('snake_right.png').convert_alpha()
        self.snake_left = pygame.image.load('snake_left.png').convert_alpha()
        
        self.tail_up = pygame.image.load('tail_up.png').convert_alpha()
        self.tail_down = pygame.image.load('tail_down.png').convert_alpha()
        self.tail_right = pygame.image.load('tail_right.png').convert_alpha()
        self.tail_left = pygame.image.load('tail_left.png').convert_alpha()
        
        self.body_vertical = pygame.image.load('body_vertical.png').convert_alpha()
        self.body_horizontal = pygame.image.load('body_horizontal.png').convert_alpha()
        
        self.body_tr = pygame.image.load('body_tr.png').convert_alpha()
        self.body_tl = pygame.image.load('body_tl.png').convert_alpha()
        self.body_br = pygame.image.load('body_br.png').convert_alpha()
        self.body_bl = pygame.image.load('body_bl.png').convert_alpha()
        self.sound = pygame.mixer.Sound("Bite.mpeg")
    def snake(self):
        self.update_head_graphics()
        self.update_tail_graphics()
        for index, i in enumerate(self.body):
            S=pygame.Rect( int(i.x * Block_height) , int(i.y * Block_height), 
                         Block_height, Block_height)
            if index == 0:
                screen.blit(self.head , S)
            elif index == len(self.body) - 1:
                screen.blit(self.tail, S)
            else:
                previous_block =self.body[index +1] - i
                next_block = self.body[index -1 ]- i
                if previous_block.x == next_block.x:
                    screen.blit(self.body_vertical, S)
                elif previous_block.y == next_block.y :
                    screen.blit(self.body_horizontal, S)
                else:
                    if previous_block.x == -1 and next_block.y == -1 or previous_block.y == -1 and next_block.x == -1:
                        screen.blit(self.body_tl, S)
                    elif previous_block.x == -1 and next_block.y == 1 or previous_block.y == 1 and next_block.x == -1:
                        screen.blit(self.body_bl, S)
                    elif previous_block.x == 1 and next_block.y == -1 or previous_block.y == -1 and next_block.x == 1:
                        screen.blit(self.body_tr, S)
                    elif previous_block.x == 1 and next_block.y == 1 or previous_block.y == 1 and next_block.x == 1:
                        screen.blit(self.body_br, S)    
                    
                    
    def update_head_graphics(self):
        header = self.body[1] - self.body[0]
        if header == pygame.math.Vector2(1, 0): 
            self.head = self.snake_left
        if header == pygame.math.Vector2(-1, 0):
            self.head = self.snake_right
        if header == pygame.math.Vector2(0, 1):
            self.head = self.snake_up
        if header == pygame.math.Vector2(0, -1):
            self.head = self.snake_down    
    def update_tail_graphics(self):
        tailer = self.body[-2] - self.body[-1]
        if tailer == pygame.math.Vector2(1, 0): 
            self.tail = self.tail_left
        if tailer == pygame.math.Vector2(-1, 0):
            self.tail = self.tail_right
        if tailer == pygame.math.Vector2(0, 1):
            self.tail = self.tail_up
        if tailer == pygame.math.Vector2(0, -1):
            self.tail = self.tail_down 
        
    def move(self):
        if self.new_block == True:
            body_copy = self.body[:]
            body_copy.insert(0, body_copy[0] + self.direction)
            self.body= body_copy[:]
            self.new_block = False
        else:
            body_copy = self.body[:-1]
            body_copy.insert(0, body_copy[0] + self.direction)
            self.body= body_copy[:]
    def add_block(self):
        self.new_block = True
    def play_sound(self):
        self.sound.play()
    def reset(self):
        self.body = [pygame.math.Vector2(5, 10), pygame.math.Vector2(4, 10), pygame.math.Vector2(3, 10)]
        self.direction = pygame.math.Vector2(0, 0)
class Apple:
    def __init__(self):
        self.change_fruit_position()
    def Rect(self):
        Rect=pygame.Rect( int(self.position.x * Block_height) , int(self.position.y * Block_height), 
                         Block_height, Block_height)
        screen.blit(FRUIT, Rect)
    def change_fruit_position(self):
        self.x= random.randint(0, Block_width-1)
        self.y= random.randint(0, Block_width-1)
        self.position = pygame.math.Vector2(self.x, self.y)    
class Main:
    def __init__(self):
        self.snake=Snake()
        self.apple=Apple()
    def update(self):
        self.snake.move()
        self.snake_eats_apple()
        self.snake_hits()
    def snake_apple(self):
        self.grass()
        self.apple.Rect()  
        self.snake.snake()
        self.draw_score()
    def snake_eats_apple(self):
        if self.apple.position == self.snake.body[0]:
            self.apple.change_fruit_position()
            self.snake.add_block()
            self.snake.play_sound()
        for i in self.snake.body[1:]:
            if i ==  self.apple.position:
                self.apple.change_fruit_position()
    def snake_hits(self):
        if not 0 <= self.snake.body[0].x < Block_width or not 0 <= self.snake.body[0].y < Block_width:
            self.game_over()
        for i in self.snake.body[1:]:
            if i == self.snake.body[0]:
                self.game_over()
    def game_over(self):
        self.snake.reset()
    def grass(self):
        grass_color = (118,238,0)
        for i in range(Block_height):
            if i%2==0:
                for j in range(Block_height):
                    if j %2 == 0:
                        grass_rect = pygame.Rect(j*Block_height, i*Block_height , Block_height, Block_height)
                        pygame.draw.rect(screen, grass_color, grass_rect)
            else:
                for j in range(Block_height):
                    if j %2 != 0:
                        grass_rect = pygame.Rect(j*Block_height, i*Block_height , Block_height, Block_height)
                        pygame.draw.rect(screen, grass_color, grass_rect)
                
    def draw_score(self):
        score = str(len(self.snake.body) - 3)
        score_surface = FONT.render(score, True, (0,0,0))
        score_x = int(Block_height * Block_width - 60)
        score_y = int(Block_height * Block_width - 60)
        score_rect = score_surface.get_rect(center = (score_x, score_y))
        apple_rect = FRUIT.get_rect(midright = (score_rect.left, score_rect.centery))
        bg_rect = pygame.Rect(apple_rect.left, apple_rect.top,apple_rect.width + score_rect.width + 6, apple_rect.height)
        pygame.draw.rect(screen, (167, 209,61), bg_rect)
        pygame.draw.rect(screen, (0, 0, 0), bg_rect, 2)
        screen.blit(score_surface, score_rect)
        screen.blit(FRUIT, apple_rect)
SCREEN_UPDATE = pygame.USEREVENT
pygame.time.set_timer(SCREEN_UPDATE, 150)
screen = pygame.display.set_mode((Block_height * Block_width , Block_height * Block_width))
pygame.display.set_caption("Snake Game 🐍")
Main = Main()

clock = pygame.time.Clock()
while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            # quit() is opposite of init()
            pygame.quit()
            # sys is used incase if there's any problem with pygame.quit()
            sys.exit()
        if event.type == SCREEN_UPDATE:
            Main.update() 
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_UP:
                if Main.snake.direction.y !=1:
                    Main.snake.direction = pygame.math.Vector2(0,-1)
            if event.key == pygame.K_DOWN:
                if Main.snake.direction.y !=-1:
                    Main.snake.direction = pygame.math.Vector2(0,1)
            if event.key == pygame.K_RIGHT:
                if Main.snake.direction.x !=-1:
                    Main.snake.direction = pygame.math.Vector2(1,0) 
            if event.key == pygame.K_LEFT:
                if Main.snake.direction.x !=1:
                    Main.snake.direction = pygame.math.Vector2(-1,0)     
    screen.fill((127,255,0)) 
    Main.snake_apple()
    pygame.display.update()
    # To run the game consistenly and not to run faster than the computer
    clock.tick(60)
    
```
